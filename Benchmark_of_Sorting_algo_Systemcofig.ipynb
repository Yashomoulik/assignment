```python
def Insertion_Sort(A):
    for i in range(1, len(A)):
        x = A[i]
        j = i - 1
        while j >= 0 and x < A[j]:
            A[j + 1] = A[j]
            j -= 1
            A[j + 1] = x
    return A
Sorted_Array = Insertion_Sort([8,45,10,9,70])
print(Sorted_Array)
```

```python
def Selection_Sort(A):
    for i in range(len(A)):
        m = i
        for j in range(i + 1, len(A)):
            if A[j] < A[m]:
                m = j
        A[i], A[m] = A[m], A[i]
    return A
Sorted_Array = Selection_Sort([8,45,10,9,70])
print(Sorted_Array)
```

```python
def Bubble_Sort(A):
    for i in range(len(A)):
        swapped = False
        for j in range(0, len(A) - i - 1):
            if A[j] > A[j + 1]:
                A[j], A[j + 1] = A[j + 1], A[j]
                swapped = True
        if not swapped:
            break
    return A
Sorted_Array = Bubble_Sort([8,45,10,9,70])
print(Sorted_Array)
```

```python
import timeit
import numpy as np
import matplotlib.pyplot as plt

def benchmark(func, sizes):
    times = []
    for size in sizes:
        arr = np.random.randint(0, 1000, size)
        start_time = timeit.default_timer()
        func(arr.copy())
        end_time = timeit.default_timer()
        times.append(end_time - start_time)
    return times

sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
insertion_times = benchmark(Insertion_Sort, sizes)
selection_times = benchmark(Selection_Sort, sizes)
bubble_times = benchmark(Bubble_Sort, sizes)

plt.plot(sizes, insertion_times, label='Insertion Sort')
plt.plot(sizes, selection_times, label='Selection Sort')
plt.plot(sizes, bubble_times, label='Bubble Sort')
plt.xlabel('Input size (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.show()
```
